```python
#3. Handling Categorical Data : -
#Categorical data means object type data , object means string type data
# Machine understand only numeric data , not understand categorical data/
#non-Numeric data 
#Means to concerts categorical type data to Numeric type data 
#How to this
```

```python
df.info()
```

```python
#Separate the numerical data and categorical data 
df_num=df.select_dtypes(["int64","float64"])
df_cat=df.select_dtypes("object")
```

```python
df_num.head()
```

```python
df_cat.head()
```

```python
#There are 2 methods , 
#to used converts categorical type data to Numeric type data 
'''1. One Hot Encoding     2. Label Encoding '''
```

```python
#1. One Hot Encoding  
df_cat["model"].unique()
```

```python
#how many gas type car and diesel type car
df_cat["model"].value_counts()
```

```python
#apply one hot encoding
pd.get_dummies(df_cat["model"]) #get_dummies() inbuilt method of pandas
```

```python
#2. Label Encoder Method
from sklearn.preprocessing import LabelEncoder
#here sklearn inbuilt package and preprocessing : inbuilt class
```

```python
#create object of LabelEncoder class
le=LabelEncoder()
```

```python
le.fit_transform(df_cat["model"]) 
#fit_transform( ) inbuilt method of LabelEncoder
```

```python
#using for loop we convert all object tyoe data into numeric type data 
for col in df_cat: 
    #create object of LabelEncoder class
    le=LabelEncoder()
    df_cat[col]=le.fit_transform(df_cat[col]) 
```
