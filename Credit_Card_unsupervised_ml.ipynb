```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("credit_card.csv")
```

```python
df
```

```python
#because of "C" CUST_ID	 is show to object we can replace "C" to ' ' 
df['CUST_ID']=pd.to_numeric(df['CUST_ID'].str.replace('C','').str.replace(',', ''), errors='coerce')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# missing value treatment for CREDIT_LIMIT
df["CREDIT_LIMIT"].fillna(df["CREDIT_LIMIT"].mean(),inplace=True)

# missing value treatment for MINIMUM_PAYMENTS
df["MINIMUM_PAYMENTS"].fillna(df["MINIMUM_PAYMENTS"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.cluster import KMeans
```

```python
ke1=KMeans(n_clusters=3,
           init="k-means++",
           n_init=10,max_iter=300,
           random_state=1)
```

```python
ke1.fit(df)
```

```python
ke1.inertia_
```

```python
centroids = ke1.cluster_centers_
```

```python
centroids
```

```python
centroid_df = pd.DataFrame(centroids,columns=["CUST_ID","BALANCE","BALANCE_FREQUENCY","PURCHASES","ONEOFF_PURCHASES",
                                              "INSTALLMENTS_PURCHASES","CASH_ADVANCE","PURCHASES_FREQUENCY",
                                             "ONEOFF_PURCHASES_FREQUENCY","PURCHASES_INSTALLMENTS_FREQUENCY",
                                             "CASH_ADVANCE_FREQUENCY","CASH_ADVANCE_TRX","PURCHASES_TRX",
                                             "CREDIT_LIMIT","PAYMENTS","MINIMUM_PAYMENTS","PRC_FULL_PAYMENT","TENURE"])
```

```python
centroid_df
```

# k value and elbow methods

```python
bcss =[] #between cluster sum of square distance
for k in range(1,10):
    ke2=KMeans(n_clusters=k, init="k-means++", n_init=10, max_iter=300,random_state=1)
    ke2.fit(df)
    bcss.append(ke2.inertia_)
```

```python
#plot elbow
plt.plot(range(1,10),bcss,'bx-')
plt.title("Elbow Method")
plt.xlabel("number of clusters")
plt.ylabel("BCSS")
plt.show()
```

```python
ke3 = KMeans(n_clusters=5, init="k-means++", n_init=10, max_iter=300,random_state=1)
```

```python
#we got common between data on based of clusters
labels = ke3.fit_predict(df)
```

```python
#calculate inertia
ke3.inertia_
```

```python
centroids = ke3.cluster_centers_
centroids
```

```python
centroid_df = pd.DataFrame(centroids,columns=["CUST_ID","BALANCE","BALANCE_FREQUENCY","PURCHASES","ONEOFF_PURCHASES",
                                              "INSTALLMENTS_PURCHASES","CASH_ADVANCE","PURCHASES_FREQUENCY",
                                             "ONEOFF_PURCHASES_FREQUENCY","PURCHASES_INSTALLMENTS_FREQUENCY",
                                             "CASH_ADVANCE_FREQUENCY","CASH_ADVANCE_TRX","PURCHASES_TRX",
                                             "CREDIT_LIMIT","PAYMENTS","MINIMUM_PAYMENTS","PRC_FULL_PAYMENT","TENURE"])
```
