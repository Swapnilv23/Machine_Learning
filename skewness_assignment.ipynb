```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Find the skewness in the following  dataset : 
```

```python
#create dictionary
skewness ={'Class Marks':[61,64,67,70,73],'Frequency':[5,18,42,27,8]}
```

```python
#converts skewness dictionary into DataFrame
df=pd.DataFrame(skewness)
print(df)
```

```python
df.columns
```

```python
df.info()
```

```python
#Handling skewness
df.columns
#find skewness of each numerical column
#then use inbuilt function skew() : it is defined skew inbuilt class
#To check skewness of each numerical column
from scipy.stats import skew  #call package
#scipy inbuilt package and stats : inbuilt class
for col in df:
    print(col)
    print(skew(df[col]))
```

```python
for col in df:
    print(col,skew(df[col]))
    plt.figure()
    sns.distplot(df[col])
    plt.axvline(df[col].mean(),color="red")
    plt.axvline(df[col].median(),color="yellow")
    plt.show()
```

```python
#to check correlation :-
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```
