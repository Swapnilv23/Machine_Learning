```python
# import required modules 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report 
```

```python
df=pd.read_csv("heart.csv")
```

```python
df
```

```python
df.info()
```

```python
x=df.iloc[:,:-1] # excluding price_range other are feature 
y=df.iloc[:,-1] # price range 
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.3)
```

# simple linear support vector

```python
from sklearn.svm import LinearSVC 
```

```python
lsv=LinearSVC(random_state=1) # becasue the line will be created randomly 
```

```python
lsv.fit(x_train,y_train)
```

```python
y_pred=lsv.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
### soft MArgine 
lsv2=LinearSVC(C=1,random_state=1)
lsv2.fit(x_train,y_train)
y_pred=lsv2.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
# wx+b 
lsv2.coef_
```

```python
lsv2.intercept_
```

# apply other kernels

```python
from sklearn.svm import SVC 
```

```python
svc=SVC(random_state=1,kernel="poly") # decision boundry will be polynomial
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
print(classification_report(y_test,y_pred))
```

# gridsearch svc

```python
from sklearn.model_selection import GridSearchCV
```

```python
C=[1.0,0.25,0.5,0.75] # margine 
kernel=["linear","rbf","poly"]
decision_function_shape=["ovo","ovr"]
# ovr= one vs rest 
# ovo = one vs one 
svmg=SVC(random_state=1)
grid_svm=GridSearchCV(estimator=svmg,
                      cv=5,
                      param_grid=dict(kernel=kernel,C=C,
                                      decision_function_shape=decision_function_shape)
                      ,verbose = 5)
grid_svm.fit(x_train,y_train)
print("best score: ", grid_svm.best_score_)
print("best param: ", grid_svm.best_params_)
#so it is also gives cross fold validation so my model will be trained with all possible 
#patterns 
```
